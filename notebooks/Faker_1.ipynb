```python
from datetime import datetime, timedelta
from random import randint, random, uniform
from faker import Faker

fake = Faker()
```

```python
_run_dt = datetime.now().astimezone()
numdays = 41

dates = [(_run_dt - timedelta(days=x)).date() for x in range(numdays, 0, -1)]
```

```python
for _ in range(5):
    print(fake.pyfloat(positive=True, left_digits=6, right_digits=4))
```

```python
fiat = fake.currency_code()
cryp = fake.cryptocurrency_code()
ticker = f"{fiat}-{cryp}"

date_from = fake.date()
date_thru = fake.date()

print(ticker, date_from, date_thru)
```

```python
tickers = ["BTC-CAD", "BTC-USD"]
rows = []
for dt in dates:
    for ticker in tickers:
        opn = uniform(0, 1000000)
        lo = uniform(0, 1000000)
        hi = uniform(0, 1000000)
        cls = uniform(0, 1000000)
        vol = randint(0, 10000000)
        row = (ticker, dt, opn, lo, hi, cls, vol)
        rows.append(row)
rows
```
